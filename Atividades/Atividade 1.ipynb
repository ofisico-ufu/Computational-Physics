```python
import numpy as np           # Para matemática numérica e vetores
import matplotlib.pyplot as plt  # Para gráficos
import math                  # Para funções matemáticas básicas (se quiser ser tradicional)
import sympy as sp 

# Exercício 1
r0_inicial = 5
a = 0.1
b = 0.9
max_iter = 1000
tol = 1e-2

# Simulação com a = 0.1
r0 = r0_inicial
trajetoria_a = []

for i in range(max_iter):
    grad_U = 2 * r0 - 1
    r = r0 - a * grad_U
    trajetoria_a.append(r)
    
    if abs(r) < tol:
        print(f"Convergência com a = {a} em {i+1} passos.")
        break
    
    r0 = r
else:
    print(f"Máximo de {max_iter} iterações atingido com a = {a}.")

# Simulação com b = 0.5
r0 = r0_inicial
trajetoria_b = []

for i in range(max_iter):
    grad_U = 2 * r0 - 1
    r = r0 - b * grad_U
    trajetoria_b.append(r)
    
    if abs(r) < tol:
        print(f"Convergência com b = {b} em {i+1} passos.")
        break
    
    r0 = r
else:
    print(f"Máximo de {max_iter} iterações atingido com b = {b}.")

# Curva do potencial
r_min = min(min(trajetoria_a), min(trajetoria_b)) - 1
r_max = max(max(trajetoria_a), max(trajetoria_b)) + 1
r_lista = np.linspace(r_min, r_max, 400)
U_lista = r_lista**2 - 1

# Gráfico
plt.figure(figsize=(10, 6))
plt.plot(r_lista, U_lista, label='Potencial U(r) = r² - 1', color='gray', linewidth=2)
plt.plot(trajetoria_a, [r**2 - 1 for r in trajetoria_a], 'o-', label='Trajetória com a = 0.1', color='blue')
plt.plot(trajetoria_b, [r**2 - 1 for r in trajetoria_b], 's--', label='Trajetória com b = 0.9', color='red')

# Estética
plt.title('Comparação de trajetórias no potencial U(r) = r² - 1')
plt.xlabel('r')
plt.ylabel('U(r)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Exercício 2

r0_inicial = 2
a = 0.05
b = 0.09
max_iter = 1000
tol = 1e-2

# Simulação com a = 0.5
r0 = r0_inicial
trajetoria_a = []

for i in range(max_iter):
    grad_U = 4 * r0**3 - 2 * r0  # Derivada de U(x) = x^4 - x^2
    r = r0 - a * grad_U
    trajetoria_a.append(r)

    grad_U_novo = 4 * r**3 - 2 * r
    if abs(grad_U_novo) < tol:
        print(f"Convergência com a = {a} em {i+1} passos.")
        break

    r0 = r
else:
    print(f"Máximo de {max_iter} iterações atingido com a = {a}.")

# Simulação com b = 0.9
r0 = r0_inicial
trajetoria_b = []

for i in range(max_iter):
    grad_U = 4 * r0**3 - 2 * r0
    r = r0 - b * grad_U
    trajetoria_b.append(r)

    grad_U_novo = 4 * r**3 - 2 * r
    if abs(grad_U_novo) < tol:
        print(f"Convergência com b = {b} em {i+1} passos.")
        break

    r0 = r
else:
    print(f"Máximo de {max_iter} iterações atingido com b = {b}.")

# Curva do potencial: U(x) = x^4 - x^2
r_min = min(min(trajetoria_a), min(trajetoria_b)) - 1
r_max = max(max(trajetoria_a), max(trajetoria_b)) + 1
r_lista = np.linspace(r_min, r_max, 400)
U_lista = r_lista**4 - r_lista**2

# Gráfico
plt.figure(figsize=(10, 6))
plt.plot(r_lista, U_lista, label='Potencial U(r) = x²(x-1)(x+1)', color='gray', linewidth=2)
plt.plot(trajetoria_a, [x**4 - x**2 for x in trajetoria_a], 'o-', label=f'Trajetória com a = {a}', color='blue')
plt.plot(trajetoria_b, [x**4 - x**2 for x in trajetoria_b], 's--', label=f'Trajetória com b = {b}', color='red')

# Estética
plt.title('Comparação de trajetórias no potencial U(r) = x²(x-1)(x+1)')
plt.xlabel('r')
plt.ylabel('U(r)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Exercício 3

r0_inicial = 2
a = 0.11
b = 0.01
max_iter = 1000
tol = 1e-2

# Simulação com a = 0.11
r0 = r0_inicial
trajetoria_a = []

for i in range(max_iter):
    grad_U = 4 * r0**3 - 2 * r0 + 0.25  # Derivada de U(x) = x^4 - x^2 + x/4
    r = r0 - a * grad_U
    trajetoria_a.append(r)

    grad_U_novo = 4 * r**3 - 2 * r + 0.25
    if abs(grad_U_novo) < tol:
        print(f"Convergência com a = {a} em {i+1} passos.")
        break

    r0 = r
else:
    print(f"Máximo de {max_iter} iterações atingido com a = {a}.")

# Simulação com b = 0.01
r0 = r0_inicial
trajetoria_b = []

for i in range(max_iter):
    grad_U = 4 * r0**3 - 2 * r0 + 0.25
    r = r0 - b * grad_U
    trajetoria_b.append(r)

    grad_U_novo = 4 * r**3 - 2 * r + 0.25
    if abs(grad_U_novo) < tol:
        print(f"Convergência com b = {b} em {i+1} passos.")
        break

    r0 = r
else:
    print(f"Máximo de {max_iter} iterações atingido com b = {b}.")

# Curva do potencial: U(x) = x^4 - x^2 + x/4
r_min = min(min(trajetoria_a), min(trajetoria_b)) - 1
r_max = max(max(trajetoria_a), max(trajetoria_b)) + 1
r_lista = np.linspace(r_min, r_max, 400)
U = r_lista**4 - r_lista**2 + r_lista / 4

# Gráfico
plt.figure(figsize=(10, 6))
plt.plot(r_lista, U, label='Potencial U(r) = x⁴ - x² + x/4', color='gray', linewidth=2)
plt.plot(trajetoria_a, [x**4 - x**2 + x / 4 for x in trajetoria_a], 'o-', label=f'Trajetória com a = {a}', color='blue')
plt.plot(trajetoria_b, [x**4 - x**2 + x / 4 for x in trajetoria_b], 's--', label=f'Trajetória com b = {b}', color='red')

# Estética
plt.title('Comparação de trajetórias no potencial U(r) = x⁴ - x² + x/4')
plt.xlabel('r')
plt.ylabel('U(r)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Exercício 4

alpha = 0.05
max_iter = 200
tol = 1e-4

# Ponto inicial
x0 = 2.5
y0 = 2.5

trajetoria = [(x0, y0)]
lista = []  # ← Aqui guardamos os valores de U(x, y) em cada iteração

# Gradiente descendente (sem função)
for i in range(max_iter):
    # Gradiente de U(x, y) = sin(x)cos(y) + 2(xy)^2 / 1000
    dU_dx = np.cos(x0) * np.cos(y0) + (4 * x0 * y0**2) / 1000
    dU_dy = -np.sin(x0) * np.sin(y0) + (4 * y0 * x0**2) / 1000

    # Atualiza posição
    x0 = x0 - alpha * dU_dx
    y0 = y0 - alpha * dU_dy

    # Valor da função
    U_xy = np.sin(x0) * np.cos(y0) + 2 * (x0 * y0)**2 / 1000
    lista.append(U_xy)  # ← Aqui armazenamos o valor de U na lista

    # Guarda posição
    trajetoria.append((x0, y0))

    # Critério de parada
    if np.sqrt(dU_dx**2 + dU_dy**2) < tol:
        print(f"Convergiu em {i+1} iterações.")
        break

trajetoria = np.array(trajetoria)

# Grade para o gráfico de contorno
x = np.linspace(-4, 4, 400)
y = np.linspace(-4, 4, 400)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y) + 2 * (X * Y)**2 / 1000

# a) Gráfico de contorno + trajetória
plt.figure(figsize=(10, 6))
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.plot(trajetoria[:, 0], trajetoria[:, 1], 'r.-', label='Trajetória')
plt.scatter(trajetoria[0, 0], trajetoria[0, 1], color='white', label='Início')
plt.scatter(trajetoria[-1, 0], trajetoria[-1, 1], color='black', label='Fim')
plt.title('Trajetória no campo de contorno de U(x, y)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.colorbar(label='U(x, y)')
plt.grid(True)
plt.show()

# b) Gráfico do valor de U(x, y) por iteração
plt.figure(figsize=(8, 4))
plt.plot(lista, 'b.-')  # ← Aqui usamos a lista para mostrar a evolução de U
plt.title('Valor de U(x, y) ao longo das iterações')
plt.xlabel('Iteração')
plt.ylabel('U(x, y)')
plt.grid(True)
plt.show()
```
